```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
TrainFile = pd.read_csv('trainRegression.csv')
TestFile = pd.read_csv('testRegression.csv')


#reading data
trainX_data=TrainFile['X']
trainY_data=TrainFile['R']

testX_data=TestFile['X']
testY_data=TestFile['R']
```

```python
#converting data in numpy Matrices

X_data_train=np.array(trainX_data)
Y_data_train=np.array(trainY_data)

X_data_test=np.array(testX_data)
Y_data_test=np.array(testY_data)

```

```python
#---Plotting data points 
plt.figure(1)
print("Training Data Plotting")
plt.plot(X_data_train,Y_data_train)
```

```python
plt.figure(2)
print("Test Data Plotting")
plt.plot(X_data_test,Y_data_test)
```

```python
#linear Regression

A=np.zeros((2,2))
B=np.arange(2)
X=np.arange(2)
```

```python
A[0][0]=X_data_train.shape[0]
A[0][1]=np.sum(X_data_train)
A[1][0]=A[0][1]
A[1][1]=np.sum(X_data_train*X_data_train)
B[0]=np.sum(Y_data_train)
B[1]=np.sum(Y_data_train*X_data_train)

#--inverse of A and then multiplied by B

inverse_A=np.linalg.inv(A)
X=np.dot(inverse_A,B)
```

```python
m=X[1]
c=X[0]

predictions=np.array(c+m*X_data_test)
```

```python
FinalPlot=plt.figure(2)
print("Linear Regression Plotting")
plt.plot(X_data_test,Y_data_test)
plt.plot(X_data_test,predictions)
```

```python
#calculating mean square
meanSquare=(np.sum((predictions-Y_data_test)*(predictions-Y_data_test))/(2*X_data_test.shape[0]))
print("Mean Square :",meanSquare)
```
