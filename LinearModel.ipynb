```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
def readDataFromCSVFile (file_name):
  data = pd.read_csv(file_name)
  return data
```

```python
TrainFile = readDataFromCSVFile('trainRegression.csv')
TestFile = readDataFromCSVFile('testRegression.csv')
```

```python
#reading data using helper function 

trainX_data=TrainFile['X']
trainY_data=TrainFile['R']

testX_data=TestFile['X']
testY_data=TestFile['R']
```

```python
#Converting data above in the form of Matrices (using Numpy)

X_data_train = np.array(trainX_data)
Y_data_train = np.array(trainY_data)

X_data_test = np.array(testX_data)
Y_data_test = np.array(testY_data)

```

```python
# Plot Training Data
plt.figure(1)
plt.plot(X_data_train,Y_data_train)
```

```python
# Plot Testing Data
plt.figure(2)
plt.plot(X_data_test,Y_data_test)
```

```python
#Main Work of Linear Regression

A = np.zeros((2,2))
B = np.arange(2)
X = np.arange(2)
```

```python

A[0][0] = X_data_train.shape[0]
A[0][1] = np.sum(X_data_train)
A[1][0] = A[0][1]
A[1][1] = np.sum(X_data_train * X_data_train)
B[0] = np.sum(Y_data_train)
B[1] = np.sum(Y_data_train * X_data_train)

# (A ^ -1) * B

inverse_A=np.linalg.inv(A)
X=np.dot(inverse_A,B)
```

```python
m = X[1]
c = X[0]

# y = mx + c
predictions = np.array(c + m*X_data_test)
```

```python
FinalPlot = plt.figure(2)
print("Linear Regression")
plt.plot(X_data_test, Y_data_test)
plt.plot(X_data_test, predictions)
```

```python
#calculating mean square
mean_square=(np.sum((predictions - Y_data_test) * (predictions - Y_data_test)) / (2 * X_data_test.shape[0]))
print("Mean Square :",mean_square)
```
